```python
import pandas as pd
import numpy as np
from scipy.stats import spearmanr
from scipy.stats import pearsonr, spearmanr

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import matplotlib as mpl

mpl.font_manager.fontManager.addfont('./THSarabunChula-Regular.ttf')
mpl.rc('font',family='TH Sarabun Chula')
```

```python
df = pd.read_excel("./Dataset_spf.xlsx")
print(f"จำนวนแถวเริ่มต้น: {len(df)}")
```

```python
df.info()
```

```python
x = pd.to_numeric(df['Attendance'], errors='coerce')
y = pd.to_numeric(df['Exam_Score'], errors='coerce')

mask = ~(x.isna() | y.isna())
x, y = x[mask], y[mask]

rho, p = spearmanr(x, y, alternative='two-sided')

# แสดงผล p-value ให้เข้าใจ (รองรับเลขเล็กมาก)
MIN_P = np.finfo(float).tiny  # 2.225e-308
p_text = f"{p:.3e}" if p > 0 else f"< {MIN_P:.3e}"

print(f"n = {len(x)}  Spearman rho = {rho:.4f}  p-value = {p_text}")
```

```python
# เลือกคอลัมน์ Attendance และ Exam_Score
x = pd.to_numeric(df['Attendance'], errors='coerce')
y = pd.to_numeric(df['Exam_Score'], errors='coerce')

# กรองค่า NA
mask = ~(x.isna() | y.isna())
x, y = x[mask], y[mask]

# Pearson Correlation
pearson_corr, pearson_p = pearsonr(x, y)

# Spearman Correlation
spearman_corr, spearman_p = spearmanr(x, y)

# แสดงผล
print(f"Pearson correlation: {pearson_corr:.4f}, p-value: {pearson_p:.3e}")
print(f"Spearman correlation: {spearman_corr:.4f}, p-value: {spearman_p:.3e}")
```

```python
plt.figure(figsize=(8, 6))

sns.regplot(x=x, y=y, scatter_kws={'alpha':0.6}, line_kws={'color':'red'})

plt.title('Scatter Plot of Exam Score vs Attendance')
plt.xlabel('Attendance')
plt.ylabel('Exam Score')
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# เลือกคอลัมน์ Internet_Access
# แปลงเป็น 0/1 (yes=1, no=0)
inet = df['Internet_Access'].astype(str).str.strip().str.lower().map({'yes':1, 'no':0})
exam = pd.to_numeric(df['Exam_Score'], errors='coerce')

mask = ~(inet.isna() | exam.isna())
inet, exam = inet[mask], exam[mask]

# สรุป mean/median/sd/n ของแต่ละกลุ่ม
group_stats = exam.groupby(inet).agg(['mean','median','std','count'])
group_stats.index = group_stats.index.map({0:'No', 1:'Yes'})
print(group_stats)

# ผลต่างค่าเฉลี่ยและมัธยฐาน
delta_mean   = group_stats.loc['Yes','mean']   - group_stats.loc['No','mean']
delta_median = group_stats.loc['Yes','median'] - group_stats.loc['No','median']
print({'ΔMean': delta_mean, 'ΔMedian': delta_median})
```

```python
plot_df = pd.DataFrame({
    'Internet_Access': inet.map({0: 'No', 1: 'Yes'}), # ใช้การ map กลับไปเป็น Yes/No
    'Exam_Score': exam
})
plt.figure(figsize=(8, 6))
# ใช้ sns.boxplot
sns.boxplot(x='Internet_Access', y='Exam_Score', data=plot_df,
            order=['Yes', 'No'], # กำหนดลำดับของกลุ่ม
            palette='Set2')

plt.title('Box Plot of Exam Score by Internet Access')
plt.xlabel('Internet Access')
plt.ylabel('Exam Score')
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python

# ตัวอย่าง: Motivation_Level ถ้าเก็บเป็นชื่อกลุ่ม/เรตติ้งข้อความ
tq = df['Motivation_Level'].astype(str).str.strip()
exam = pd.to_numeric(df['Exam_Score'], errors='coerce')
mask = ~(tq.isna() | exam.isna())
tq, exam = tq[mask], exam[mask]

# สรุปแต่ละกลุ่ม
by_group = exam.groupby(tq).agg(['mean','median','std','count']).sort_values('median', ascending=False)
print(by_group)

# รายงานช่วงต่าง (ระหว่างกลุ่มที่ median สูงสุด vs ต่ำสุด)
range_mean   = by_group['mean'].iloc[0]   - by_group['mean'].iloc[-1]
range_median = by_group['median'].iloc[0] - by_group['median'].iloc[-1]
print({'range_mean': range_mean, 'range_median': range_median})
```

```python
order = ['High', 'Medium', 'Low']

plt.figure(figsize=(10, 6))
# สร้าง DataFrame ชั่วคราวสำหรับการพล็อต
plot_df = pd.DataFrame({'Motivation_Level': tq, 'Exam_Score': exam})

# ใช้ sns.boxplot
sns.boxplot(x='Motivation_Level', y='Exam_Score', data=plot_df,
            order=order,
            palette='Pastel2')

plt.title('Box Plot of Exam Score by Motivation Level')
plt.xlabel('Motivation Level')
plt.ylabel('Exam Score')
plt.grid(axis='y', linestyle='--')
plt.show()
```
